

#### Input/Output

* `input()` → Konsoldan (terminaldan) qiymet almaq ucun istifade edilir

  * `input()` funksiyasinin qiymeti `str` tipinde olur
  * `a1 = input()`
  * `print(a1)`

  **İzah**: `input()` istifadəçidən məlumat alır və onu `str` (string) formatında geri qaytarır.

---

* `print()` → Konsola (terminala) qiymet capa vermek ucun istifade edilir

  * `print()` funksiyasinda stringler 3 formada formatlanir:

---

**1. `format()` funksiyasindan istifade ederek:**

* `ad = input()`
* `soyad = input()`
* və ya
* `ad, soyad = input().split(" ")`
* `print("{} {}".format(ad, soyad))` → və ya
* `print("{0} {1}".format(ad, soyad))`

**İzah**: `format()` funksiyası dəyişənləri müəyyən bir yerə daxil edərək çap edir. `{}` istifadə olunur və həmin yerdə dəyişənlər yerləşdirilir.

---

**2. `%` simvolundan istifade ederek:**

* `print("%s %s" % (ad, soyad))`
* `print("%d %d" % (ad, soyad))`
* `print("%f %f" % (ad, soyad))`

**İzah**: `%` operatoru, formatlama əməliyyatını həyata keçirir. `%s` simvolu stringləri, `%d` rəqəmləri və `%f` ondalıklı ədədləri formatlamaq üçün istifadə olunur.

---

**3. `f` simvolundan istifade ederek:**

* `print(f"{ad} {soyad}")`

**İzah**: `f` formatlı simvollarla, dəyişənləri birbaşa formatlanmış cümləyə daxil etmək mümkündür. `f"{}"` sintaksisindən istifadə edilir.

---

* `print()` funksiyasinin `end` və `sep` parametrleri var:

* `print(ad, end=" ")` → sonda bosluq qoyur

* `print(soyad, end=",")` → sonda vergul qoyur

* `print(ad, soyad, sep=", ")` → deyisenler arasinda vergul ve bosluq qoyur

**İzah**:

* `end`: Çapın sonuna nə əlavə edəcəyinizi göstərir. `end=" "` boşluq qoyur, `end=","` vergul qoyur və s.
* `sep`: Dəyişənlər arasında nə qoyulacağını göstərir. `sep=", "` vergul və boşluq qoyar.
