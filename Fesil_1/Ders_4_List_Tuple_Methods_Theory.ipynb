## List Functions



* `append()` → listin axırına `append()`-ə yazdığımız qiyməti əlavə edir və `None` qaytarır
  → `list1.append(4.5)`

* `remove()` → listdən `remove()`-a yazdığımız qiyməti silirik və `None` qaytarır
  → `list1.remove(4)`

* `insert()` → listin neçənci element sırasına qiymət daxil etmək üçün birinci indeksi, sonra qiyməti yazırıq və `None` qaytarır
  → `list1.insert(5, "salam")`

* `pop()` → default vəziyyətdə `pop()` listin axırıncı elementini silir və onun qiymətini qaytarır
  → `list1.pop()`
  → `list1.pop(3)`

* `extend()` → `extend()`-ə daxil etdiyimiz `iterable` tipli dəyişəni string kimi parçalayaraq listin axırına ayrı-ayrı elementlər kimi əlavə edir və `None` qaytarır
  → `list1.extend("53")`

* `count()` → `count()`-a daxil etdiyimiz dəyərin baş vermə sayını qaytarır və listə heç bir dəyişiklik etmir
  → `k = list1.count(50)`

* `index()` → `index()`-ə daxil etdiyimiz dəyərin indeksini qaytarır və listə heç bir dəyişiklik etmir
  → `k = list1.index(-21)`

* `reverse()` → `reverse()` listi tərsinə çevirir və `None` qaytarır
  → `list1.reverse()`

* `sort()` → `sort()` listi azdan çoxa, `sort(reverse=True)` isə çoxdan aza sıralayır və `None` qaytarır
  → `list1.sort()`
  → `list1.sort(reverse=True)`

* `clear()` → `clear()` listin elementlərini tamamilə silir və `None` qaytarır
  → `list1.clear()`

* `copy()` → `copy()` listin nüsxəsini yaradır və yeni bir dəyişənə yazılmalıdır
  → `list2 = list1.copy()`
  
---  

* listin elementini dəyişmək üçün → məsələn: `list[3] = 9` → listdə 3-cü elementin qiyməti 9 ilə əvəz olunur

* listdə indeksi yazıb qiyməti çap etmək üçün → məsələn: `print(list[5])` → 5-ci elementin qiymətini çap edir

* `iterable` tipli dəyişənin elementlərinin sayı (uzunluğu) → məsələn: `print(len(tuple103))`
