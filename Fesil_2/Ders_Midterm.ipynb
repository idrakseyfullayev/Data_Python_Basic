### **English question**

- ### Write a program that reads multi-digit numbers until 0 and calculates the sum of the digit 2 of the input numbers.

### **Azerbaycanca**

- ### 0 daxil edilənə qədər çoxrəqəmli ədədləri oxuyun və daxil edilən bütün ədədlərin **2 rəqəminin cəmini** hesablayın.

# **2. Task**

### **English question**
- ### Write a program that reads numbers separated by spaces until EOF and counts the number of the input numbers that end in 5, line by line.

### **Azerbaycanca**
- ### Sətir-sətir verilmiş, aralarında boşluq olan ədədləri oxuyun və hər sətirdə **5 rəqəmi ilə bitən** ədədlərin sayını çap edin.


````python
while True:
    try:
        line = input()
    except EOFError:
        break

    nums = line.split()
    count = 0
    for n in nums:
        if n.endswith("5"):
            count += 1
    print(count)

````

# **3. Task**

### **English question**

- ### Write a program that reads strings until the string "END" and replace each consonant with an "*" character.

### **Azerbaycanca**

- ### "END" sözü daxil edilənə qədər bütün sözləri oxuyun və bütün **samit hərfləri** `*` işarəsi ilə əvəz edin.




````python
vowels = "aeiouAEIOU"

while True:
    s = input()
    if s == "END":
        break

    res = ""
    for ch in s:
        if ch.isalpha() and ch not in vowels:
            res += "*"
        else:
            res += ch
    print(res)
````

# **4. Task**

### **English question**

- ### Write a program that randomly generates **n integers** between 10 and 70 and determines the **geometric mean** of the numbers divisible by 3.
- ### Print the result to three decimal places.
- ### If there are no such numbers, print 0.

### **Azerbaycanca**

- ### 10 ilə 70 arasında təsadüfi olaraq **n ədəd tam ədəd** yaradın. İçərisində **3-ə bölünən** ədədlərin **həndəsi ortasını** hesablayın.
- ### Nəticəni 3 onluq dəqiqliklə çap edin.
- ### Əgər belə ədəd yoxdursa, 0 çıxarın.

````python
    import random
    import math

    n = int(input("n = "))
    nums = [random.randint(10, 70) for _ in range(n)]
    print("Generated:", *nums)

    filtered = [x for x in nums if x % 3 == 0]

    if len(filtered) == 0:
        print(0)
    else:
        product = 1
        for x in filtered:
            product *= x
        geo_mean = product ** (1/len(filtered))
        print(f"{geo_mean:.3f}")
````

# **1. Task

- ### **English:**
- ### Write a program that reads multi-digit numbers until 0 and returns the product of the even digits of the input number.

- ### **Azerbaijani:**
- ### 0-a qədər çoxrəqəmli ədədləri oxuyan və daxil edilən ədədlərin cüt rəqəmlərinin hasilini qaytaran proqram yazın.

````python
result = 1
while True:
    n = int(input())
    if n == 0:
        break

    product = 1
    has_even = False
    for digit in str(n):
        if int(digit) % 2 == 0:
            product *= int(digit)
            has_even = True

    if has_even:
        result *= product
    else:
        result *= 0  
print(result)
````

# 2. Task

- ### **English:**
- ### Write a program that reads numbers separated by spaces until the EOF and counts the elements which are square numbers (1, 4, 9, 16, 25, …).

- ### **Azerbaijani:**
- ### Mətn sonuna qədər (EOF) boşluqla ayrılmış ədədləri oxuyan və kvadrat ədədlərin sayını hesablayan proqram yazın.

````python
import math
while True:
    line = input()
    if line == "0":  
        break
    
    count = 0
    nums = map(int, line.split())
    for n in nums:
        if int(math.sqrt(n)) ** 2 == n:
            count += 1

    print(count)
````

# **4. Task

- ### **English:**
- ### Write a program that randomly generates n elements between 15 and 65, counts the arithmetic mean of the even values, and displays the result with two decimal places.

- ### **Azerbaijani:**
- ### 15 və 65 arasında n element təsadüfi yaradan, cüt ədədlərin arifmetik ortasını hesablayan və nəticəni iki onluq yerlə göstərən proqram yazın.

- ### **English:**
- ### Write a program that reads multi-digit numbers (≥10) until 0 and returns the sum of the digits of the smallest input number.

- ### **Azerbaijani:**
- ### 0-a qədər çoxrəqəmli ədədləri (≥10) oxuyan və daxil edilmiş ədədlərdən ən kiçiyinin rəqəmlərinin cəmini qaytaran proqram yazın.

- ### Write a program that reads three ascending numbers separated by spaces until EOF and prints YES if they form a -### Pythagorean triplet (`a² + b² = c²`) and NO otherwise.

- ### **Azerbaijani:**
- ### EOF-a qədər artan sıra ilə 3 ədədi oxuyan və əgər ədədlər Pifaqor üçlüyünü (`a² + b² = c²`) təşkil edirsə YES, əks halda NO çap edən proqram yazın.

````python
while True:
    line = input()
    if line == "0":  
        break

    a, b, c = line.split()
    if int(a)**2 + int(b)**2 == int(c)**2:
        print("YES")
    else:
        print("NO")
````

# **1. Task**

- ### **English:**
- ### Write a program that reads multi-digit numbers until 0 and calculates the difference between the sum of even and odd digits of the input numbers.

- ### **Azerbaijani:**
- ### 0 daxil edilənədək çoxrəqəmli ədədləri oxuyun və həmin ədədlərin cüt rəqəmlərinin cəmi ilə tək rəqəmlərinin cəminin fərqini hesablayın.



````python
while True:
    n = int(input())
    if n == 0:
        break

    even_sum = 0
    odd_sum = 0

    for d in str(n):
        d = int(d)
        if d % 2 == 0:
            even_sum += d
        else:
            odd_sum += d

    print(even_sum - odd_sum)
````

# **4.Task

- ### **English:**

- ### Write a program that randomly generates **n integers between 10 and 100**, prints these numbers, finds the three numbers divisible by 3, calculates their geometric mean (( (a \cdot b \cdot c)^{1/3} )), and prints it to three decimal places. If fewer than 3 such numbers exist, print 0.

### **Azerbaijani:**

- ### 10 və 100 arasında **n ədəd** təsadüfi yaradan, onları ekrana çıxaran, 3-ə bölünən ilk üç ədədi seçən, onların həndəsi ortasını (( (a \cdot b \cdot c)^{1/3} )) hesablayan və üç onluqla çap edən proqram yazın. Əgər belə ədəd 3-dən azdırsa, 0 yazın.

````python
import random
n = int(input("n = "))
nums = [random.randint(10, 100) for _ in range(n)]
print(*nums)

div3 = [x for x in nums if x % 3 == 0]

if len(div3) >= 3:
    a, b, c = div3[:3]
    geo = (a * b * c) ** (1/3)
    print(f"{geo:.3f}")
else:
    print(0)
````
