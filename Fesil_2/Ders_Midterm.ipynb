# **3. Task**

### **English question**

- ### Write a program that reads strings until the string "END" and replace each consonant with an "*" character.

### **Azerbaycanca**

- ### "END" sözü daxil edilənə qədər bütün sözləri oxuyun və bütün **samit hərfləri** `*` işarəsi ilə əvəz edin.
**Input:**
```
apple
pear
END
```
**Output:**
```
a***e
*ea*
```

````python
vowels = "aeiouAEIOU"
list_1 = []
while True:
    s = input()
    if s == "END":
        break

    res = ""
    for ch in s:
        if ch.isalpha() and ch not in vowels:
            res += "*"
        else:
            res += ch
    list_1.append(res)
    
for i in list_1:    
    print(i)
````

### **English question**

- ### Write a program that randomly generates **n integers** between 10 and 70 and determines the **geometric mean** of the numbers divisible by 3.
- ### Print the result to three decimal places.
- ### If there are no such numbers, print 0.

### **Azerbaycanca**

- ### 10 ilə 70 arasında təsadüfi olaraq **n ədəd tam ədəd** yaradın. İçərisində **3-ə bölünən** ədədlərin **həndəsi ortasını** hesablayın.
- ### Nəticəni 3 onluq dəqiqliklə çap edin.
- ### Əgər belə ədəd yoxdursa, 0 çıxarın.
**Input:**
```
n=5
```
**Output:**
```
21 14 46 60 42
3.544
```

````python
    import random
    import math

    n = int(input("n = "))
    nums = [random.randint(10, 70) for _ in range(n)]
    print(*nums)

    filtered = [x for x in nums if x % 3 == 0]

    if len(filtered) == 0:
        print(0)
    else:
        product = 1
        for x in filtered:
            product *= x
        geo_mean = product ** (1/len(filtered))
        print(f"{geo_mean:.3f}")
````

- ### **Azerbaijani:**

```

```

````python
import math
list_1 = []
while True:
    line = input()
    if line == "0":  
        break
    
    count = 0
    nums = map(int, line.split())
    for n in nums:
        if int(math.sqrt(n)) ** 2 == n:
            count += 1
    list_1.append(count)
    
for i in list_1:    
    print(i)
````

# ✅ 4) Task
- ### English:
- ### Write a program that randomly generates n elements between 25 and 85, and calculates the average of the odd values. Display the result with three decimal places.

- ### Azerbaijani:
- ### n sayda ədədi 25 ilə 85 arasında təsadüfi seçən və tək ədədlərin orta qiymətini hesablayan proqram yazın. Nəticəni üç onluq dəqiqliklə çap edin.
**Input:**
```
n = 5
21 22 46 61 45
```
**Output:**
```
42.33
```

````python
import random

n = int(input("n = "))
sum_odd = 0
count_odd = 0
odd_values = []

for _ in range(n):
    num = random.randint(25, 85)
    if num % 2 != 0:
        sum_odd += num
        count_odd += 1
        odd_values.append(num)

print("Odd values:", *odd_values)
if count_odd > 0:
    print("Average: {:.3f}".format(sum_odd / count_odd))
else:
    print("No odd values generated")
````

````python
import random

n = int(input())
numbers = [random.randint(25, 85) for _ in range(n)]
odds = [x for x in numbers if x % 2 != 0]
print(*odds)

if odds:
    mean = sum(odds) / len(odds)
    print(f"{mean:.3f}")
else:
    print("0")
````
