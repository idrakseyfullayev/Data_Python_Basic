# Şəkil 1
## **2. Task**

### **English question**
- ### Write a program that reads numbers separated by spaces until EOF and counts the number of the input numbers that end in 5, line by line.

### **Azerbaycanca**
- ### Sətir-sətir verilmiş, aralarında boşluq olan ədədləri oxuyun və hər sətirdə **5 rəqəmi ilə bitən** ədədlərin sayını çap edin.
**Input:**
```
25 16 5 36
23
45 7 85 15
0
```
**Output:**
```
2
0
3
```

### sys.stdin görə ancaq Visual Stuido Code-da işləyir - input()-ı əvəz edir

````python
import sys
list_1 = []
for line in sys.stdin:
    nums = line.split()
    count = 0
    for n in nums:
        if n.endswith("5"):
            count += 1
    list_1.append(count)
    
for i in list_1:    
    print(i)
````

# Şəkil 2
## 2. Task

- ### **English:**
- ### Write a program that reads numbers separated by spaces until the EOF and counts the elements which are square numbers (1, 4, 9, 16, 25, …).

- ### **Azerbaijani:**
- ### Mətn sonuna qədər (EOF) boşluqla ayrılmış ədədləri oxuyan və kvadrat ədədlərin sayını hesablayan proqram yazın.
**Input:**
```
2 16 5 36
25
4 7 81 1
0
```
**Output:**
```
2
1
3
```

### sys.stdin görə ancaq Visual Stuido Code-da işləyir - input()-ı əvəz edir

````python
import sys
import math
list_1 = []
for line in sys.stdin:
    nums = map(int, line.split())
    count = 0
    for n in nums:
        if int(math.sqrt(n)) ** 2 == n:
            count += 1
    list_1.append(count)
    
for i in list_1:    
    print(i)
````

# Şəkil 3
## **2. Task**

- ### **English:**
- ### Write a program that reads three ascending numbers separated by spaces until EOF and prints YES if they form a -### Pythagorean triplet (`a² + b² = c²`) and NO otherwise.

- ### **Azerbaijani:**
- ### EOF-a qədər artan sıra ilə 3 ədədi oxuyan və əgər ədədlər Pifaqor üçlüyünü (`a² + b² = c²`) təşkil edirsə YES, əks halda NO çap edən proqram yazın.
**Input:**
```
3 4 5
2 6 8
7 24 25
0
```
**Output:**
```
YES
NO
YES
```

### sys.stdin görə ancaq Visual Stuido Code-da işləyir - input()-ı əvəz edir

````python
import sys
list_1 = []
for line in sys.stdin:
    result = ""
    a, b, c = map(int, line.split())
    if a**2 + b**2 == c**2:
        result = "YES"
    else:
        result = "NO"
    list_1.append(result)
    
for i in list_1:
    print(i)
````

# Şəkil 4
## **2. Task**

- ### **English:**

- ### Write a program that reads integers separated by spaces until EOF and calculates the sum of the elements divisible by 9 in each row.

- ### **Azerbaijani:**

- ### EOF-a qədər boşluqla ayrılmış tam ədədləri oxuyun və hər sətirdə 9-a bölünən ədədlərin cəmini hesablayın.
**Input:**
```
45  2  25  9  4
7
9 18
0
```
**Output:**
```
54
0
27
```

### sys.stdin görə ancaq Visual Stuido Code-da işləyir - input()-ı əvəz edir

````python
import sys
list_1 = []
for line in sys.stdin:
    nums = map(int, line.split())
    s = sum(x for x in nums if x % 9 == 0)
    list_1.append(s)
    
for i in list_1:    
    print(i)
````

- ### **English:**

- ### Write a program that reads multi-digit numbers until the EOF and counts the number of digit **'4'** in each input number.

- ### **Azerbaijani:**

- ###  EOF-a qədər çoxrəqəmli ədədləri oxuyun və hər bir ədəddəki **'4' rəqəmlərinin sayını** hesablayın.
**Input:**

```
142341
482
524436
```
**Output:**
```
2
1
2
```

### sys.stdin görə ancaq Visual Stuido Code-da işləyir - input()-ı əvəz edir

````python
import sys
list_1 = []
for line in sys.stdin:
    list_1.append(line.count("4"))
    
for i in list_1:
    print(i)
````

### sys.stdin görə ancaq Visual Stuido Code-da işləyir - input()-ı əvəz edir

````python
import sys
list_1 = []
for line in sys.stdin:
    count_4 = 0
    for d in line.strip():
        if d == '4':
            count_4 += 1
    list_1.append(count_4)

for i in list_1: 
    print(i)    
````
