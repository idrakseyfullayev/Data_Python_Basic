## Class

Əlbəttə, aşağıda hər bir anlayışın **Python class** kontekstində **bir cümlə ilə** tərifini verirəm:

1. **Static variable** – Bütün instanslar (obyektlər) tərəfindən **ortaq istifadə olunan**, class səviyyəsində təyin edilən dəyişəndir.
2. **Dynamic variable** – Hər instans üçün **ayrı-ayrılıqda** yaranan və `__init__` daxilində və ya sonradan obyektə əlavə olunan dəyişəndir.
3. **`__init__` method** – Class-dan yeni obyekt yaradıldıqda **ilk çağırılan və obyektin başlanğıc vəziyyətini təyin edən** xüsusi metoddur.
4. **Inheritance** – Bir class-ın başqa bir class-dan **xüsusiyyət və metodları miras almasına** deyilir.
5. **Inheritance növləri** – Python-da **Single**, **Multiple**, **Multilevel**, **Hierarchical** və **Hybrid inheritance** dəstəklənir.
6. **Polymorphism** – Eyni adlı metodların **fərqli class-larda fərqli şəkildə işləməsi** xüsusiyyətidir.
7. **Abstraction** – İstifadəçiyə **yalnız lazım olan detalları göstərib**, daxili işləmə qaydasını gizlətmək prinsipidir (`abc` modulu ilə).
8. **Encapsulation** – Məlumatların **xüsusi (private)** saxlanması və yalnız metodlar vasitəsilə **nəzarətli girişə** icazə verilməsi prinsipidir.
9. **Decorator** – Mövcud funksiyanın **davranışını dəyişdirmək və ya genişləndirmək** üçün istifadə olunan xüsusi funksiyalardır (`@` ilə yazılır).
10. **Generator** – `yield` açarı ilə yazılmış və **tənbəl şəkildə dəyər qaytaran**, iterator yaradan funksiyalardır.

## Decoratorlar

### 1️⃣ `@abstractmethod`

**Tərif**: Əsas class-da müəyyən metodun alt class-lar tərəfindən mütləq təyin edilməsini tələb edən abstrakt metoddur.
**Nümunə**:

```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Woof")
```

---

### 2️⃣ `@classmethod`

**Tərif**: Class səviyyəsində işləyən, `cls` parametrini qəbul edən və class dəyişənlərinə çıxış imkanı verən metoddur.
**Nümunə**:

```python
class User:
    count = 0
    def __init__(self):
        User.count += 1

    @classmethod
    def total(cls):
        print(cls.count)
```

---

### 3️⃣ `@staticmethod`

**Tərif**: Nə `self`, nə də `cls` istifadə etməyən, class daxilində ancaq məntiqi aidiyyəti olan müstəqil funksiya kimi işləyən metoddur.
**Nümunə**:

```python
class Math:
    @staticmethod
    def add(a, b):
        return a + b
```

---

### 4️⃣ `@property`

**Tərif**: Metodu bir atribut kimi istifadə etməyə imkan verən, getter funksiyası üçün dekoratordur.
**Nümunə**:

```python
class Circle:
    def __init__(self, r):
        self._r = r

    @property
    def area(self):
        return 3.14 * self._r ** 2
```

---

### 5️⃣ `@decorator`

**Tərif**: Başqa funksiyaları qəbul edib onların davranışını dəyişən və ya genişləndirən funksiyadır.
**Nümunə**:

```python
def my_decorator(func):
    def wrapper():
        print("Start")
        func()
        print("End")
    return wrapper

@my_decorator
def say_hello():
    print("Hello")
```

---

### 6️⃣ `Generator`

**Tərif**: `yield` açar sözü ilə nəticəni bir-bir qaytaran və yaddaş sərfini azaldan xüsusi funksiya növüdür.
**Nümunə**:

```python
def my_gen():
    yield 1
    yield 2
    yield 3

for i in my_gen():
    print(i)
```
